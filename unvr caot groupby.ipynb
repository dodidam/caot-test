### 1. Import library and load dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('unvr CAOT - output.csv')
df.head(5)
```

---
### 2. Groupby exclude

```python
excluded = df.groupby(by='Exclude',dropna=False).count()[['Company Code']]
excluded.rename(columns={'Company Code':'Count of Billing Document'},inplace=True)
excluded.loc['Grand Total'] = excluded['Count of Billing Document'].sum()
excluded
```

---
### 3. Groupby channel

```python
df_filter = df[(df['Exclude'].isna() == True) & (df['Ageing'] != '>2')]
channel = df_filter.groupby(by='Channel').count()[['Ageing']]
channel.rename(columns={'Ageing':'Count of Billing Document'},inplace=True)
channel.loc['Grand Total'] = channel['Count of Billing Document'].sum()
channel
```
